```python
# pip install rdkit-pypi pandas numpy
from __future__ import annotations

from dataclasses import dataclass
from typing import Dict, List, Optional, Tuple

import numpy as np
import pandas as pd

from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem
from rdkit.Chem.Scaffolds import MurckoScaffold


def bemis_murcko_scaffold(smiles: str, include_chirality: bool = False) -> Optional[str]:
    if not isinstance(smiles, str) or not smiles.strip():
        return None
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None
    try:
        scaf = MurckoScaffold.GetScaffoldForMol(mol)
        if scaf is None:
            return None
        return Chem.MolToSmiles(scaf, isomericSmiles=include_chirality)
    except Exception:
        return None


def morgan_fp(mol: Chem.Mol, radius: int = 2, nbits: int = 2048) -> DataStructs.ExplicitBitVect:
    # ECFP4 default = radius=2
    return AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=nbits)


def bulk_max_similarity(query_fp, fps_list) -> float:
    # Max tanimoto similarity of query_fp against fps_list
    if not fps_list:
        return 0.0
    sims = DataStructs.BulkTanimotoSimilarity(query_fp, fps_list)
    return float(max(sims)) if sims else 0.0


def scaffold_representative_fp(
    fps: List[DataStructs.ExplicitBitVect],
    rng: np.random.Generator,
    method: str = "medoid",
    max_medoid_size: int = 200,
) -> DataStructs.ExplicitBitVect:
    """
    Return one representative fp for a scaffold.
    method:
      - "medoid": choose fp with highest mean similarity to others (robust representative)
      - "random": random fp
    max_medoid_size: if scaffold is huge, subsample for medoid computation.
    """
    if len(fps) == 1:
        return fps[0]

    if method == "random":
        return fps[int(rng.integers(0, len(fps)))]

    if method != "medoid":
        raise ValueError("method must be 'medoid' or 'random'")

    # Subsample if large to keep it fast
    if len(fps) > max_medoid_size:
        idx = rng.choice(len(fps), size=max_medoid_size, replace=False)
        fps_sub = [fps[i] for i in idx]
    else:
        fps_sub = fps

    # Compute mean similarity for each candidate to all others (O(k^2) within scaffold)
    best_i = 0
    best_score = -1.0
    for i, fp in enumerate(fps_sub):
        sims = DataStructs.BulkTanimotoSimilarity(fp, fps_sub)
        score = float(np.mean(sims))  # includes self-sim=1.0 (fine, constant shift)
        if score > best_score:
            best_score = score
            best_i = i
    return fps_sub[best_i]


def greedy_scaffold_count_split(
    scaf_to_indices: Dict[str, List[int]],
    target_test: int,
    seed: int = 0,
) -> Tuple[List[str], List[str]]:
    """
    Greedy packing of scaffolds into test to hit target molecule count.
    Returns (train_scaffolds, test_scaffolds).
    """
    rng = np.random.default_rng(seed)
    items = list(scaf_to_indices.items())
    rng.shuffle(items)
    items.sort(key=lambda kv: len(kv[1]), reverse=True)

    test_scafs: List[str] = []
    train_scafs: List[str] = []
    test_count = 0

    for scaf, idxs in items:
        g = len(idxs)
        dist_if_test = abs((test_count + g) - target_test)
        dist_if_train = abs(test_count - target_test)

        if dist_if_test < dist_if_train:
            test_scafs.append(scaf)
            test_count += g
        elif dist_if_test > dist_if_train:
            train_scafs.append(scaf)
        else:
            if rng.random() < 0.5:
                test_scafs.append(scaf)
                test_count += g
            else:
                train_scafs.append(scaf)

    return train_scafs, test_scafs


def objective_mean_test_to_train_maxsim(
    test_reps: List[DataStructs.ExplicitBitVect],
    train_reps: List[DataStructs.ExplicitBitVect],
) -> float:
    """
    Objective to MINIMIZE: mean over test reps of (max similarity to any train rep).
    Lower = more OOD separation.
    """
    if not test_reps or not train_reps:
        return 1.0  # degenerate; treat as bad
    maxsims = [bulk_max_similarity(fp, train_reps) for fp in test_reps]
    return float(np.mean(maxsims))


@dataclass
class OODSplitResult:
    train_idx: np.ndarray
    test_idx: np.ndarray
    train_fraction: float
    test_fraction: float
    n_invalid_smiles: int
    n_scaffolds_train: int
    n_scaffolds_test: int
    objective: float  # mean test->train max similarity (lower is better)


def scaffold_fp_ood_split(
    df: pd.DataFrame,
    smiles_col: str = "SMILES",
    test_frac: float = 0.50,
    seed: int = 0,
    include_chirality: bool = False,
    drop_invalid: bool = True,
    fp_radius: int = 2,
    fp_nbits: int = 2048,
    rep_method: str = "medoid",      # "medoid" or "random"
    max_medoid_size: int = 200,
    max_iters: int = 2000,
    patience: int = 400,
    verbose: bool = True,
) -> OODSplitResult:
    """
    Scaffold-disjoint split with local-search optimization to maximize fingerprint separation.
    - Maintains scaffold disjointness.
    - Targets ~50/50 by molecule count (exact as possible via scaffold grouping).
    - Optimizes to minimize mean(test max similarity to train).
    """
    if smiles_col not in df.columns:
        raise KeyError(f"'{smiles_col}' not found in df columns.")
    if not (0.0 < test_frac < 1.0):
        raise ValueError("test_frac must be in (0, 1).")

    rng = np.random.default_rng(seed)

    smiles = df[smiles_col].tolist()
    mols = []
    scaffolds = []
    invalid = []
    for s in smiles:
        scaf = bemis_murcko_scaffold(s, include_chirality=include_chirality)
        m = Chem.MolFromSmiles(s) if isinstance(s, str) else None
        ok = (scaf is not None) and (m is not None)
        invalid.append(not ok)
        mols.append(m if ok else None)
        scaffolds.append(scaf if ok else None)

    invalid_mask = np.array(invalid, dtype=bool)
    n_invalid = int(invalid_mask.sum())
    if n_invalid and not drop_invalid:
        raise ValueError(f"Found {n_invalid} invalid SMILES. Set drop_invalid=True to drop them.")

    valid_idx = np.where(~invalid_mask)[0]
    n_total = len(valid_idx)
    if n_total == 0:
        raise ValueError("No valid molecules after filtering invalid SMILES.")

    # Group indices by scaffold
    scaf_to_indices: Dict[str, List[int]] = {}
    for i in valid_idx:
        scaf_to_indices.setdefault(scaffolds[i], []).append(int(i))

    # Compute fps per molecule, grouped per scaffold
    scaf_to_fps: Dict[str, List[DataStructs.ExplicitBitVect]] = {}
    for scaf, idxs in scaf_to_indices.items():
        fps = []
        for i in idxs:
            fp = morgan_fp(mols[i], radius=fp_radius, nbits=fp_nbits)
            fps.append(fp)
        scaf_to_fps[scaf] = fps

    # Representative fp per scaffold
    scafs = list(scaf_to_indices.keys())
    scaf_rep_fp: Dict[str, DataStructs.ExplicitBitVect] = {}
    for scaf in scafs:
        scaf_rep_fp[scaf] = scaffold_representative_fp(
            scaf_to_fps[scaf], rng=rng, method=rep_method, max_medoid_size=max_medoid_size
        )

    target_test = int(round(test_frac * n_total))

    # Initial scaffold packing to hit target_test
    train_scafs, test_scafs = greedy_scaffold_count_split(scaf_to_indices, target_test, seed=seed)

    def count_mols(scaf_list: List[str]) -> int:
        return sum(len(scaf_to_indices[s]) for s in scaf_list)

    # Repair counts a bit: move smallest scaffolds until closer to target
    def repair_counts(train_scafs: List[str], test_scafs: List[str]) -> Tuple[List[str], List[str]]:
        changed = True
        while changed:
            changed = False
            tc = count_mols(test_scafs)
            dist_now = abs(tc - target_test)

            if tc > target_test:
                # move smallest from test to train if it helps
                test_scafs.sort(key=lambda s: len(scaf_to_indices[s]))
                for k, scaf in enumerate(test_scafs):
                    new_tc = tc - len(scaf_to_indices[scaf])
                    if abs(new_tc - target_test) < dist_now:
                        train_scafs.append(scaf)
                        del test_scafs[k]
                        changed = True
                        break
            elif tc < target_test:
                train_scafs.sort(key=lambda s: len(scaf_to_indices[s]))
                for k, scaf in enumerate(train_scafs):
                    new_tc = tc + len(scaf_to_indices[scaf])
                    if abs(new_tc - target_test) < dist_now:
                        test_scafs.append(scaf)
                        del train_scafs[k]
                        changed = True
                        break
        return train_scafs, test_scafs

    train_scafs, test_scafs = repair_counts(train_scafs, test_scafs)

    # Build rep fp lists
    def reps(scaf_list: List[str]) -> List[DataStructs.ExplicitBitVect]:
        return [scaf_rep_fp[s] for s in scaf_list]

    train_reps = reps(train_scafs)
    test_reps = reps(test_scafs)

    best_obj = objective_mean_test_to_train_maxsim(test_reps, train_reps)
    best_train_scafs = list(train_scafs)
    best_test_scafs = list(test_scafs)

    if verbose:
        print(f"Initial: test mols={count_mols(test_scafs)}/{n_total} "
              f"({count_mols(test_scafs)/n_total:.2%}), objective={best_obj:.4f}")

    # Local search: swap one scaffold between train/test (keeps disjointness)
    # Accept if objective improves and test molecule count stays as close as possible to target.
    no_improve = 0
    for it in range(max_iters):
        # pick candidate scaffolds to swap
        s_test = test_scafs[int(rng.integers(0, len(test_scafs)))]
        s_train = train_scafs[int(rng.integers(0, len(train_scafs)))]

        # enforce count closeness: swapping changes test count by delta = size(train) - size(test)
        tc = count_mols(test_scafs)
        delta = len(scaf_to_indices[s_train]) - len(scaf_to_indices[s_test])
        new_tc = tc + delta

        # Only consider swaps that do not worsen distance-to-target (or improve it)
        if abs(new_tc - target_test) > abs(tc - target_test):
            continue

        # Try swap
        new_test_scafs = list(test_scafs)
        new_train_scafs = list(train_scafs)
        new_test_scafs.remove(s_test)
        new_train_scafs.remove(s_train)
        new_test_scafs.append(s_train)
        new_train_scafs.append(s_test)

        new_obj = objective_mean_test_to_train_maxsim(reps(new_test_scafs), reps(new_train_scafs))

        if new_obj < best_obj - 1e-6:
            # accept improving move
            test_scafs, train_scafs = new_test_scafs, new_train_scafs
            best_obj = new_obj
            best_test_scafs, best_train_scafs = list(test_scafs), list(train_scafs)
            no_improve = 0
            if verbose and (it % 100 == 0 or it < 50):
                print(f"iter={it}: improved objective={best_obj:.4f}, "
                      f"test mols={count_mols(best_test_scafs)} ({count_mols(best_test_scafs)/n_total:.2%})")
        else:
            no_improve += 1
            if no_improve >= patience:
                break

    # Final indices
    train_idx = np.array([i for scaf in best_train_scafs for i in scaf_to_indices[scaf]], dtype=int)
    test_idx = np.array([i for scaf in best_test_scafs for i in scaf_to_indices[scaf]], dtype=int)

    # Sanity
    assert len(set(train_idx).intersection(set(test_idx))) == 0
    assert len(train_idx) + len(test_idx) == n_total

    train_fraction = len(train_idx) / n_total
    test_fraction = len(test_idx) / n_total

    if verbose:
        print(f"Final: train={len(train_idx)} ({train_fraction:.2%}), "
              f"test={len(test_idx)} ({test_fraction:.2%}), "
              f"objective={best_obj:.4f}")
        print(f"Scaffolds: train={len(best_train_scafs)}, test={len(best_test_scafs)}")
        if n_invalid:
            print(f"Dropped invalid SMILES: {n_invalid}")

    return OODSplitResult(
        train_idx=train_idx,
        test_idx=test_idx,
        train_fraction=train_fraction,
        test_fraction=test_fraction,
        n_invalid_smiles=n_invalid,
        n_scaffolds_train=len(best_train_scafs),
        n_scaffolds_test=len(best_test_scafs),
        objective=best_obj,
    )


# ---- Example usage ----
# df = pd.read_csv("molecules.csv")  # must include smiles column
df = mlsmr_data
res = scaffold_fp_ood_split(df, smiles_col="SMILES", test_frac=0.50, seed=42, max_iters=3000, patience=600)
train_df = df.iloc[res.train_idx].reset_index(drop=True)
test_df  = df.iloc[res.test_idx].reset_index(drop=True)
print("Mean test->train max similarity (lower is more OOD):", res.objective)
```
